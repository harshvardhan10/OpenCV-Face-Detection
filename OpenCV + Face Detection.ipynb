```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('group_photo.jpeg')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
model = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces=model.detectMultiScale(img)
# Returns: No. of rows = Faces
# x,y,w,h
```

```python
x,y,w,h = faces[0]
```

```python
new_img=cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0))
```

```python
plt.imshow(new_img)
```

```python
for one_face in faces:
    x,y,w,h = one_face
    new_img=cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0))
```

```python
plt.imshow(new_img)
```
